## Task 2

In this task, we measure several statistics of the sepal lengths and save it into `iris_stats` for statistics on all the `iris.data` and `stats_0`, `stats_1`, and `stats_2` for each species using the data you have filtered in **Task 1**:

- maximum (max)

- mininum (min)

- mean

- median

- standard deviation (std)

Implement a function `sepal_length_statistics(m)`:

- `m` is a `list` of measurements

- returns a dictionary with keys `max`, `min`, `mean`, `median` and `std` that stores the statistics.

**Solution requirements:**

- round all the statistics to at most 2 decimals. *Use the function `np.around` to help you to round the answers.*

- `m` is a `list` not an NDArray

- You MUST use combinations of `map`/`reduce`/`filter`. *List comprehension* is allowed but do not overuse it - majority of the solution must come from the functional programming methods.

- Do not use for or while-loops for you solution. 

- Do not use `Pandas`.

- Do not use `NumPy` to calculate the statistics measures.
    

````python
stats_0 = [measure_0[x][0] for x in range(len(measure_0))]
stats_1 = [measure_1[x][0] for x in range(len(measure_1))]
stats_2 = [measure_2[x][0] for x in range(len(measure_2))]

def sepal_length_statistics(m):
    result = {
        'max' : sorted(m)[-1],
        'min' : sorted(m)[0],
        'mean' : np.around(reduce(lambda a, b: a + b, m) / len(m), 2),
        'median' : ((sorted(m)[int(len(m)/2)]) + (sorted(m)[int(len(m)/2) + 1])) / 2,
        'std' : np.around((sum((x - (sum(m) / len(m))) ** 2 for x in m) / len(m)) ** 0.5, 2)
    }

    return result 
    
print(f'Species 0 Stats: {sepal_length_statistics(stats_0)}')
print(f'Species 1 Stats: {sepal_length_statistics(stats_1)}')
print(f'Species 2 Stats: {sepal_length_statistics(stats_2)}')

print(f'Iris Stats: {sepal_length_statistics(iris.data[:,0])}')
````

## Task 3 

For this question, we want to examine the difference between mean and median of different features for each of three species.

Implement a function `mean_median_diff(m, feature_name)`.

- `m`: a NDArray of measurements

- `feature_name`: the feature as a string.  It must be one of `sepal length`, `sepal width`, `petal length`, or `petal width`.

- returns the absolute difference between the mean and the median of the measurements for `feature_name`.

**Solution Requirements:**

- Report the absolute value of the different with upto 4 decimals. *You can use `np.round` to perform rounding to the right number of decimal places.*

- Use the dictionary `features` with the `apply` function to identify/map the proper data column index from the given `feature_name`.

- You may, but not required to use the function from **Task 2** (`sepal_length_statistics(m)`).

- Do not use for or while-loops for you solution.


````python
features = {
    'sepal length': 0,
    'sepal width': 1,
    'petal length': 2,
    'petal width': 3,
}


def mean_median_diff(m, feature_name):
    
    index_finder = features.get(feature_name)

    mean = np.mean(m[:,index_finder])
    median = np.median(m[:,index_finder])
    
    result = np.around((abs(mean - median)),4)
    
    return result 

converted_my_stats0 = list(map(lambda x: list(x), measure_0))
converted_my_stats1 = list(map(lambda x: list(x), measure_1))
converted_my_stats2 = list(map(lambda x: list(x), measure_2))

array_0 = np.array(converted_my_stats0)
array_1 = np.array(converted_my_stats1)
array_2 = np.array(converted_my_stats2)


# Testing your solution - uncomment to run
for feature in features.keys():
    print(feature, mean_median_diff(iris.data, feature))
for feature in features.keys():    
    print(feature, mean_median_diff(array_0, feature))
for feature in features.keys():
    print(feature, mean_median_diff(array_1, feature))
for feature in features.keys():
    print(feature, mean_median_diff(array_2, feature))
````

## Task 5

In this task, we are calculating the `sepal_area` defined by a sepal's $width \times length$, based on that you will print a report for all measurements with `sepal_area` below the `sepal_area` mean for the given measurements. Your report will follow the format:

```
    sepal_area: species | sepal_length | sepal_width | petal_height | petal_width
    ...
    We found <N> measurements.
```

where `N` is the number of measurements that fitted the criteria.

For this task, you must provide a solution via 3 different methods, implement one function for each:

1. `funcReport`: Functional programming using combinations of `map`, `reduce`, `filter`, `sorted`.   

    - receives `m` a `list` of measurements, and `s` a `list` of species.
    
2. `npReport`: NumPy using NDArrays, this may include the usage *functional programming* as a helper.

    - receives `m` a NDArray of measurements, and `s` NDArray of species.
    
3. `pdReport`: Pandas using DataFrames, this may include the usage of `numpy` or *functional programming* as a helper.

    - receives `df` a DataFrame (like `df_iris`from **Task 4**)

**Do not use for or while-loops in your solutions.**

````python
# funcReport 

def multiply(x, y):
    return round((x * y), 2)

def funcReport(m, s):

    my_stats = m
    my_target = s 
    converted_my_stats = list(map(lambda x: x.tolist(), my_stats))
    converted_my_target = list(map(lambda x: x.tolist(), my_target))
    
    
    species = converted_my_target
    sepal_length = list(map(lambda x: x[0], converted_my_stats))
    sepal_width = list(map(lambda x: x[1], converted_my_stats))
    petal_length = list(map(lambda x: x[2], converted_my_stats))
    petal_width = list(map(lambda x: x[3], converted_my_stats))
    
    sepal_area = list(map(multiply, sepal_length, sepal_width))
                      
    sepal_area_mean = reduce(lambda x, y: x + y, sepal_area) / len(sepal_area)
    
    sepal_area_under_mean = list(filter(lambda x: x < sepal_area_mean, sepal_area))
    
    message = "sepal_area: %.2f species %d | sepal_length %.2f | sepal_width %.2f | petal_length %.2f |  petal_width %.2f" 
    result = [print (message % (a,b,c,d,e,f)) for(a,b,c,d,e,f) in zip(sepal_area, species,sepal_length, sepal_width, petal_length, petal_width) if a < sepal_area_mean]
    print(f'We found <{len(sepal_area_under_mean)}> measurements')
    
    
    

funcReport(iris['data'], iris['target'])
````

````python
# npReport

def npReport(m, s):
    
    sepal_area = m[:,0] * m[:,1]
    
    sepal_area_under_mean = np.where(sepal_area < np.mean(sepal_area))
    
    lst = list(filter(lambda x: x < np.mean(sepal_area), sepal_area))
    
    
    species = s
    sepal_length = m[:,0]
    sepal_width = m[:,1]
    petal_length = m[:,2]
    petal_width = m[:,3]
    
    message = "sepal_area: %.2f species %d | sepal_length %.2f | sepal_width %.2f | petal_length %.2f |  petal_width %.2f" 
    
    result = [print (message % (a,b,c,d,e,f)) for(a,b,c,d,e,f) in zip(sepal_area, species,sepal_length, sepal_width, petal_length, petal_width) if a < np.mean(sepal_area)]
    print(f'We found <{len(lst)}> measurements')

    
npReport(iris['data'], iris['target'])
````
